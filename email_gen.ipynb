```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://careers.nike.com/analyst-finance-corporate-investments-and-capital-expenditures/job/R-79091")
nike_page_data = loader.load().pop().page_content
print(nike_page_data)
```

```python
from langchain_core.prompts import PromptTemplate
promt_extract = PromptTemplate.from_template(
    """
    ### SCRAPED TEXT FROM WEBSITE:
{nike_page_data}

### INSTRUCTION:
The scraped text above is taken from the careers page of a company website.
Your task is to extract the job posting information and return it strictly in
JSON format with the following keys:

- role
- experience
- skills
- description

Guidelines:
- Infer experience if it is mentioned explicitly or implicitly (e.g., "entry-level", "2+ years").
- Combine technical and soft skills under the "skills" key as an array.
- The description should be a concise summary of the role and responsibilities.
- Do not include any extra text, explanations, or formatting outside JSON.

### VALID JSON (NO PREAMBLE):

"""
)

chain_extract = promt_extract | llm 
res = chain_extract.invoke(input={'nike_page_data':nike_page_data})
type(res.content)
```

```python
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser()
json_conv = json_parser.parse(res.content)
json_conv
```
